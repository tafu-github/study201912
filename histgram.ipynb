ヒストグラムの作成を調べた　excelよりも簡単で便利そう

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# 平均 50, 標準偏差 10 の正規乱数を1000件生成
x = np.random.normal(50, 10, 1000)
 
# ヒストグラムを出力 0-100まで(range)を　100分割(bins)で作成する
# 変数 = plt.hist()でヒストグラムを作成した棒グラフの値がタプル形式(変更不可能)で保存される

hist_data = plt.hist(x, range = (0, 100), bins = 100)
```

```python
# 棒グラフを作成しているデータ
hist_data
```

```python
# pandas形式で中身の確認(見やすいから)
import pandas as pd
hist_pd=pd.DataFrame(hist_data,index=["hindo","kukan","Rectangle"])
```

```python
# 区間の情報、区間での頻出回数、棒グラフ作成情報から構成されている
hist_pd
```

```python
# 保存
hist_pd.to_csv('hist.csv')
```

```python
# 補足：Rectangleとは何か調べてみた
# matplotlib.patches.Rectangle⇒長方形を書くためのmatplotlibのクラス
# 特別な用途がない限り使わなそう　これはexcelでいい

import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline

fig = plt.figure()
ax = plt.axes()

test1 = matplotlib.patches.Rectangle(xy=(0, 0), width=1, height=5, angle=0)
test2 = matplotlib.patches.Rectangle(xy=(2, 2), width=1, height=5, angle=0)
test3 = matplotlib.patches.Rectangle(xy=(4, 0), width=2, height=2, angle=0)
test4 = matplotlib.patches.Rectangle(xy=(6, 4), width=5, height=1, angle=45)

ax.add_patch(test1)
ax.add_patch(test2)
ax.add_patch(test3)
ax.add_patch(test4)

plt.xlim([0,10])
plt.ylim([0,10])
```
